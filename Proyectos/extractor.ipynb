```python
from datetime import datetime
import pandas as pd
import requests
from bs4 import BeautifulSoup
from io import StringIO

url = "https://www.myfxbook.com/forex-economic-calendar"
codificacion='utf-8'

response = requests.get(url)
response.encoding = codificacion
soup = BeautifulSoup(response.text, 'html.parser')

tablas = pd.read_html(StringIO(str(soup)), encoding=codificacion)
tabla = tablas[0]

tabla.rename(columns={'Unnamed: 3':'Símbolo'}, inplace=True)

tabla.drop(tabla['Unnamed: 2'].dropna().index, inplace=True)

tabla.drop(columns=['All  None', 'Unnamed: 2'], inplace=True) #para poner otros datos solo se pone columns=['','',''...]

tabla.drop(tabla[tabla['Impact'] == 'Low'].index, inplace=True)


tabla.dropna(subset='Impact', inplace=True)

tabla.reset_index(drop=True, inplace=True)

simbolos = tabla['Símbolo'].unique()

list(simbolos)
try: 
    with pd.ExcelWriter(path=f"Calendarios/Calendario_{datetime.now().strftime("%Y-%m-%d")}.xlsx") as writer:
        for i in simbolos:
            signos = tabla[tabla['Símbolo'] == i]
            signos.to_excel(writer, sheet_name=i, index=False)
except :
    for i in simbolos:
        signos = tabla[tabla['Símbolo'] == i]
        print(signos.head(5))
```
