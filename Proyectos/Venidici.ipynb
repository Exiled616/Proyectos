```python
import yfinance as yf
from datetime import datetime, timedelta, timezone
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
def cotización():
    Lista = []
    Df_lista = []
    respuesta = input("Inserte el símbolo que desea buscar")
    while '.' not in respuesta:   
        Lista.append(respuesta)
        respuesta = input("Inserte el símbolo que desea buscar")
    tiempo = int(input("¿Para cuántos días desea hacer la consulta?"))
    df = yf.download(
        Lista,
        start=datetime.today() - timedelta(days=tiempo),
        end=datetime.today(),
        auto_adjust=False,
        progress=False,
        interval='5m'
    )
    df.index = df.index.tz_convert(tz=('America/Bogota'))
    # df.columns = df.columns.droplevel(1)
    df = df[['Adj Close', 'Volume']]
    fig = go.Figure()
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    for i in Lista:
        fig.add_trace(go.Scatter(x=df.index, y=df['Adj Close'][i], mode='lines', name=i), secondary_y=False)
        fig.add_trace(go.Scatter(x=df.index, y=df['Volume'][i], mode='lines', name=i), secondary_y=True)
    fig.update_layout(title='Gráfica de Precios y Volumen', xaxis_title='Fecha', yaxis_title='Precio', yaxis2_title='Volumen')
    fig.show()
    display(df)
    
    # for j in Df_lista:
    #     pd.concat([Df_lista[j]], [Df_lista[j+1]])
    # print(Df_lista)    
if __name__ == "__main__":
    cotización()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta, timezone

# Descargar los datos de Tesla
respuesta = input('Inserte símbolos que desea consular separados por un espacio').upper()
lista = respuesta.split()
tiempo = int(input('¿Para cuántos días desea hacer la consulta?'))
intervalo = input('Que intervalo de tiempo desea consultar?')
df = yf.download(
    lista,
    start=datetime.today() - timedelta(days=tiempo),
    end=datetime.today(),
    interval=intervalo,
    progress=False,
    auto_adjust=False
)
fig = make_subplots(
    rows=1, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.1, 
    specs=[[{'secondary_y': True}]]  
)
indiceReal = df.index
indiceFalso = range(len(df.index))
df.set_index(pd.Index(indiceFalso), inplace=True)
for i in lista:
    fig.add_trace(
        go.Scatter(
            x=df.index,
            y=df['Adj Close'][i],
            mode='lines',
            name='Acciones de ' + yf.Ticker(i).info['shortName'],
        )
    )
    fig.add_trace(
        go.Bar(
            x=df.index,
            y=df['Volume'][i],
            name='Volomen de ' + yf.Ticker(i).info['shortName'],
            opacity=0.5,
        ), secondary_y=True
    )
fig.update_xaxes(
    tickvals=list(indiceFalso),
    ticktext=indiceReal.strftime('%Y/%m/%d: %H:%M'),
    dtick=10000,
)
fig.update_layout(
    title="Precio Ajustado y Volumen de: " + ', '.join([yf.Ticker(i).info['shortName'] for i in lista]) +'<br>' + 'Intervalo de tiempo: ' + intervalo,
    title_font=dict(family='Sans-Serif', size=16),
    title_font_weight='bold',
    xaxis_title="Fecha y hora",
    yaxis_title="Precio Ajustado",
    yaxis2_title="Volumen",
    legend_title="Símbolos",
)

fig.show()
```
