```python
import yfinance as yf
from datetime import datetime, timedelta, timezone
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def cotización():
    """Busca información de una empresa directamente con yfinance."""
    search_term = input("Escribe el nombre o símbolo de la empresa: ").strip()
    try:
        # Intentar buscar directamente por el símbolo o nombre
        symbol = yf.Ticker(search_term)
        info = symbol.info
        # Validar si la información está disponible
        if 'longName' in info:
            print(f"Nombre de la empresa: {info['longName']} ({info.get('symbol', 'N/A')})")
            print(
                f"{info.get('companyOfficers', [{'title': 'No disponible'}])[0]['title']}: {info.get('companyOfficers')[0]['name']}\n"
                f"Precio actual por acción: ${info.get('currentPrice', 'No disponible')}\n" 
                f"Recomendación técnica: {info.get('recommendationKey', 'No disponible')}\n"
                f"Crecimiento Trimestral: {round((info.get('earningsQuarterlyGrowth', 0)*100),2)}%\n"
                f"Márgenes operativos: {round(info.get('operatingMargins', 0)*100,2)}%\n"
                f"RoA: {round((info.get('returnOnAssets', 0)*100),2)}%\n"
                f"RoI: {round((info.get('returnOnEquity', 0)*100),2)}%\n"
                f"Valor de la empresa: $ {info.get('enterpriseValue', 0):,.2f}\n"
                f"Free Cash Flow: ${info.get('freeCashflow', 0):,.2f}\n"
                f"Página para inversionistas: {info.get('irWebsite', 'No disponible')}\n"
                
            )
            tiempo = int(input('¿Días de consulta hasta la fecha actual?'))
            df = yf.download(
                info['symbol'],
                start=datetime.today() - timedelta(days=tiempo),
                end=datetime.now(),
                auto_adjust=False,
                progress=False,
                interval="5m"
                )
            df.index = df.index.tz_convert('America/Bogota')
            # df = dfb[dfb['Volume'] != 0]
            # df = df.dropna(how='all')
            # df.to_excel(excel_writer='saramambicho.xlsx')
            fig, ax1 = plt.subplots(figsize=(12, 6))

            # Graficamos el primer conjunto de datos en el primer eje y (ax1)
            ax1.plot(df.index, df['Adj Close'], color='b', label='Precio de Cierre Ajustado')
            ax1.set_xlabel('Fecha: Bogotá, Colombia (UTC -5)')
            ax1.set_ylabel('Precio (USD)', color='b')
            ax1.tick_params(axis='y', labelcolor='b')
            
            # Creamos el segundo eje y (ax2) para manejar otra escala
            ax2 = ax1.twinx()
            
            # Supongamos que queremos graficar el volumen de operaciones en el segundo eje
            ax2.plot(df.index, df['Volume'], color='g', label='Volumen', linestyle='--')
            ax2.set_ylabel('Volumen de operaciones', color='g')
            ax2.tick_params(axis='y', labelcolor='g')
            # Título y leyenda
            plt.title(f'Gráfica de {symbol.info['shortName']} con Precio y Volumen')
            fig.tight_layout()  # Ajustamos el espacio para que no se sobrepongan las etiquetas
            plt.show()
            
        else:
            print("No se encontró información para el término ingresado.")
    except Exception as e:
        print(f"Error al buscar información: {e}")

if __name__ == "__main__":
    cotización()
```

```python
import yfinance as yf
from datetime import datetime, timedelta
import pandas as pd
import matplotlib.pyplot as plt

def cotización():
    search_term = input("Escribe el nombre o símbolo de la empresa: ").strip()
    try:
        # Intentar buscar directamente por el símbolo o nombre
        symbol = yf.Ticker(search_term)
        info = symbol.info
        # Validar si la información está disponible
        if 'longName' in info:
            print(f"Nombre de la empresa: {info['longName']} ({info.get('symbol', 'N/A')})")
            print(
                f"{info.get('companyOfficers', [{'title': 'No disponible'}])[0]['title']}: {info.get('companyOfficers')[0]['name']}\n"
                f"Precio actual por acción: ${info.get('currentPrice', 'No disponible')}\n" 
                f"Recomendación técnica: {info.get('recommendationKey', 'No disponible')}\n"
                f"Crecimiento Trimestral: {round((info.get('earningsQuarterlyGrowth', 0)*100),2)}%\n"
                f"Márgenes operativos: {round(info.get('operatingMargins', 0)*100,2)}%\n"
                f"RoA: {round((info.get('returnOnAssets', 0)*100),2)}%\n"
                f"RoE: {round((info.get('returnOnEquity', 0)*100),2)}%\n"
                f"Valor de la empresa: $ {info.get('enterpriseValue', 0):,.2f}\n"
                f"Free Cash Flow: ${info.get('freeCashflow', 0):,.2f}\n"
                f"Página para inversionistas: {info.get('irWebsite', 'No disponible')}\n"
            )
            tiempo = int(input('¿Días de consulta hasta la fecha actual? '))
            df = yf.download(
                info['symbol'],
                start=datetime.today() - timedelta(days=tiempo),
                end=datetime.now(),
                auto_adjust=False,
                progress=False,
                interval="5m"
            )
            df.index = df.index.tz_convert('America/Bogota')

            # Filtrar períodos con volumen cero y datos faltantes
            df = df[df['Volume'] != 0]
            df = df.dropna()

            # Crear índice continuo para comprimir espacios vacíos
            df['index_continuo'] = range(len(df))
            
            # Crear una columna de volumen suavizado con media móvil
            ventana_suavizado = 5  # Cambia este valor para ajustar la cantidad de suavización
            df['Volumen_Suavizado'] = df['Volume'].rolling(window=ventana_suavizado, min_periods=1).mean()

            # Graficar usando el índice continuo
            fig, ax1 = plt.subplots(figsize=(12, 6))

            # Gráfica del precio de cierre ajustado
            ax1.plot(df['index_continuo'], df['Adj Close'], color='b', label='Precio de Cierre Ajustado')
            ax1.set_ylabel('Precio (USD)', color='b')
            ax1.tick_params(axis='y', labelcolor='b')

            # Configurar el eje x para mostrar las fechas correctamente
            def format_date(x, p):
                if x < 0 or x >= len(df):
                    return ''
                return df.index[int(x)].strftime('%Y-%m-%d %H:%M')

            ax1.xaxis.set_major_formatter(plt.FuncFormatter(format_date))
            plt.xticks(rotation=45)

            # Crear el segundo eje para el volumen suavizado
            ax2 = ax1.twinx()
            ax2.plot(df['index_continuo'], df['Volumen_Suavizado'], color='g', label='Volumen Suavizado', linestyle='--')
            ax2.set_ylabel('Volumen de operaciones', color='g')
            ax2.tick_params(axis='y', labelcolor='g')

            # Título y ajustes finales
            plt.title(f'Gráfica de {info["shortName"]} con Precio y Volumen Suavizado')
            plt.grid(True, alpha=0.3)

            # Ajustar el espaciado
            fig.tight_layout()
            plt.show()
            
        else:
            print("No se encontró información para el término ingresado.")
    except Exception as e:
        print(f"Error al buscar información: {e}")

if __name__ == "__main__":
    cotización()
```
